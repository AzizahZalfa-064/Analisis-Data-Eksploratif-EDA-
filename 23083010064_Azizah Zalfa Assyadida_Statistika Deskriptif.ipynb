## Visualisasi Data

```python
# Visualisasi variabel Age
plt.figure(figsize=(7, 5))
sns.histplot(df['Age'], kde=True, bins=20, color='lightblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Histogram diatas menunjukkan bahwa variabel Age termasuk dalam distribusi normal dengan sedikit skew ke kanan. Kurva yang simetris di sekitar rata-rata usia (20-30 thn) menunjukkan bahwa sebagian besar pelanggan berusia muda, dengan sedikit penurunan jumlah pelanggan di usia yang lebih tua. Sehingga, histogram tidak membentuk lonceng sempurna seperti visualisasi distribusi normal pada umumnya. 

```python
# Visualisasi Variabel Annual Income
plt.figure(figsize=(7, 5))
sns.histplot(df['Annual Income ($)'], kde=True, bins=20, color='lightgreen')
plt.title('Distribution of Annual Income ($)')
plt.xlabel('Annual Income ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Hasil distribusi dari variabel Annual Income dapat dikategorikan kedalam distribusi Positively Skewed (Left-Skewed) karena terlihat menunjukkan skew kiri yang kuat. Dengan mayoritas pelanggan memiliki pendapatan kurang dari $100,000 menyebabkan ekor panjang di salah satu sisinya.

Dapat dilihat bahwa histogram diatas menunjukkan hasil yang merata di seluruh rentang nilai antara 1-100. Ini berarti tidak ada kecenderungan khusus dari poin pengeluaran pelanggan. Karenanya, variabel Spending Score termasuk dalam distribusi Uniform.

```python
# Visualisasi variabel Work Experience
plt.figure(figsize=(7, 5))
sns.histplot(df['Work Experience'], kde=True, bins=20, color='orange')
plt.title('Distribution of Work Experience')
plt.xlabel('Work Experience (years)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Selain Annual Income, variabel Work Experience juga menunjukkan adanya skew yang jelas pada histogram.
```python
# Visualisasi variabel Family Size
plt.figure(figsize=(7, 5))
sns.histplot(df['Family Size'], kde=True, bins=20, color='purple')
plt.title('Distribution of Family Size')
plt.xlabel('Family Size')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Histogram hasil visualisasi variabel Family Size termasuk dalam kategori distribusi Multimodal karena terlihat memiliki beberapa puncak yang mencerminkan adanya kelompok keluarga yang lebih umum ditemukan. ## Pemusatan Data

```python
kolom_numerik = ['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size']
hasil = []

for perhitungan in kolom_numerik:
    mean = df[perhitungan].mean()
    median = df[perhitungan].median()
    modus = df[perhitungan].mode()[0]
    
    hasil.append({
        'Faktor Pengaruh': perhitungan,
        'Mean': mean,
        'Median': median,
        'Modus': modus
    })
    
hasil_df = pd.DataFrame(hasil)
print(hasil_df)
```

Pemusatan data dapat dilakukan dengan mencari mean, median, dan modus dari data. Berikut penjelasan mengenai hasil perhitungan dari pemusatan data yang saya dapatkan

1. Age : Sebagian besar pelanggan berada pada rentang usia dewasa, dengan usia yang paling banyak ditemukan adalah 31, meskipun terdapat banyak variasi usia dari pelanggan lebih tua yang menyebabkan kenaikan rata-rata sampai ke angka 48.

2. Annual Income : Mean dan median dari variabel ini memiliki nilai yang serupa, menunjukkan bahwa distribusi pendapatan cukup simetris. Namun, modus dari data ini adalah $9,000, menunjukkan pelanggan berpenghasilan cukup rendah memiliki jumlah paling banyak.

3. Spending Score : Poin dari skor pengeluaran pelanggan memiliki tingkat distribusi yang merata dan simetris. Dengan mean, median, dan modus yang bernilai serupa.

4. Work Experience : Sebagian besar pelanggan memiliki pengalaman kerja yang singkat dilihat dari hasil perhitungan modus. Dengan nilai rata-rata lebih tinggi karena adanya beberapa pelanggan yang berpengalaman lebih lama.

5. Family Size : Sebagian besar pelanggan memiliki keluarga berukuran sedang (3-4 orang). Tetapi keluarga dengan anggota kecil (2 orang) adalah yang paling banyak ditemui.

## Penyebaran Data

```python
kolom_numerik = ['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size']
hasil = []

for perhitungan in kolom_numerik:
    std_dev = df[perhitungan].std()
    varian = df[perhitungan].var()
    rentang = df[perhitungan].max() - df[perhitungan].min()
    q1 = df[perhitungan].quantile(0.25)
    q2 = df[perhitungan].median()
    q3 = df[perhitungan].quantile(0.75)
    
    hasil.append({
        'Faktor Pengaruh': perhitungan,
        'Standar Deviasi': std_dev,
        'Varians': varian,
        'Rentang': rentang,
        'Q1': q1,
        'Q2': q2,
        'Q3': q3
    })
    
hasil_df = pd.DataFrame(hasil)
print(hasil_df)
```

Hasil perhitungan diatas menunjukkan bahwa Variabel Age tersebar luas dengan standar deviasi yang tinggi, dengan sebagian besar pelanggan berusia 25 hingga 73 tahun. Begitu pula dengan Annual Income, persebaran luas menunjukkan adanya perbedaan besar antara jumlah pelanggan berpenghasilan rendah dan tinggi. Sedangkan variabel Spending Score tersebar merata mencakup seluruh rentang dari 1 hingga 100, mencerminkan poin perilaku belanja yang beragam.  Untuk variabel Work Experience, memiliki penyebaran yang cukup rendah dengan mayoritas pelanggan memiliki pengalaman kerja kurang dari 7 tahun. Di sisi lain, Family Size relatif stabil, dengan mayoritas pelanggan memiliki 25 anggota keluarga.

## Analisis Skewness

```python
from scipy.stats import skew

kolom_numerik = ['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size']

for column in kolom_numerik:
    skewness_value = skew(df[column])
    
    print(f'Skewness of {column}: {skewness_value}')
```

berikut adalah penjelasan mengenai hasil analisis skewness 


1. Age (Skewnes 0.049): Skewness mendekati 0 menunjukkan distribusi yang hampir simetris, dengan sedikit kecenderungan ke arah kanan. Menunjukkan usia pelanggan hampir terdistribusi  secara merata.

2. Annual Income ($) (Skewness -0.116): Skewness dengan nilai negatif menunjukkan distribusi dari variabel ini cenderung ke arah kiri. Menunjukkan bahwa pelanggan dengan pendapatan lebih besar berjumlah sedikit lebih banyak walau distribusinya hampir simetris.

3. Spending Score (1-100) (Skewness 0.004): Skewness dengan nilai hampir 0 menunjukkan distribusi yang hampir sempurna simetris. Ini berarti poin dari skor pengeluaran pelanggan tersebar merata.

4. Work Experience (Skewness 0.683): Skewness dengan nilai positif menunjukkan skew ke arah kanan. Artinya mayoritas pelanggan hanya memiliki sedikit pengalaman kerja. 

5. Family Size (Skewness 0.199): Skewness dengan nilai positif kecil menunjukkan distribusi variabel ini mengarah sedikit ke kanan, di mana mayoritas pelanggan memiliki ukuran keluarga yang kecil hingga sedang.

Selanjutnya adalah visualisasi dari hasil analisis skewness menggunakan KDE. 

### Visualisasi Analisis Skewness

```python
kolom_numerik = ['Age', 'Annual Income ($)', 'Spending Score (1-100)', 'Work Experience', 'Family Size']

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))
axes = axes.flatten()

for i, perhitungan in enumerate(kolom_numerik):
    sns.kdeplot(df[perhitungan], color='pink', fill=True, alpha=0.5, ax=axes[i])
    
    mean = df[perhitungan].mean()
    median = df[perhitungan].median()
    modus = df[perhitungan].mode()[0]
    skewness = skew(df[perhitungan])
    
    axes[i].axvline(mean, color='blue', linestyle='-', linewidth=2, label=f'Mean: {mean:.2f}')  # Garis mean
    axes[i].axvline(median, color='red', linestyle=':', linewidth=3, label=f'Median: {median:.2f}')  # Garis median
    axes[i].axvline(modus, color='green', linestyle='-.', linewidth=2, label=f'Modus: {modus:.2f}')  # Garis modus
    
    axes[i].set_title(f'Distribution of {perhitungan} (Skewness: {skewness:.2f})', fontsize=12)
    axes[i].set_xlabel(perhitungan, fontsize=10)
    axes[i].set_ylabel('Density', fontsize=10)
    
    if skewness > 0:
        print(f"{perhitungan}: Distribusi condong ke kanan (right-skewed).")
    elif skewness < 0:
        print(f"{perhitungan}: Distribusi condong ke kiri (left-skewed).")
    else:
        print(f"{perhitungan}: Distribusi simetris.")

if len(kolom_numerik) < len(axes):
    for j in range(len(kolom_numerik), len(axes)):
        fig.delaxes(axes[j])  

plt.tight_layout()
plt.show()
```
